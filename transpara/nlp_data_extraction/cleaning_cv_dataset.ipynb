```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
resumeDataSet = pd.read_csv('data/UpdatedResumeDataSet.csv' ,encoding='utf-8')
resumeDataSet['cleaned_resume'] = ''
resumeDataSet.head()
```

```python
resumeDataSet.info()
```

```python
print ("Displaying the distinct categories of resume:\n\n ")
print (resumeDataSet['Category'].unique())
```

```python
print ("Displaying the distinct categories of resume and the number of records belonging to each category:\n\n")
print (resumeDataSet['Category'].value_counts())
```

```python
import seaborn as sns
plt.figure(figsize=(20,5))
plt.xticks(rotation=90)
ax=sns.countplot(x="Category", data=resumeDataSet)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.01 , p.get_height() * 1.01))
plt.grid()
```

```python
resumeDataSet.isnull().sum()
```

```python
# Analysis 1: Category distribution visualization
plt.figure(figsize=(12,6))
sns.countplot(y='Category', data=resumeDataSet, order=resumeDataSet['Category'].value_counts().index)
plt.title('Resume Category Distribution')
plt.xlabel('Count')
plt.ylabel('Category')
plt.show()
```
